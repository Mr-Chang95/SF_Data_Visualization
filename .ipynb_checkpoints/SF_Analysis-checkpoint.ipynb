## Cleaning Data

```python
# Letmake a copy of the table first
clean_sf_df = sf_df.copy()
clean_sf_df.head()
```

### Define - Problem 1

**1. Rename the `NAICS Code Description` to `Industry` and the last column to `Boundaries`.**

### Code

```python
clean_sf_df = clean_sf_df.rename(columns={"DBA Name": "Name","NAICS Code Description": "Industry", "Source Zipcode": "Zipcode"
                   ,"Neighborhoods - Analysis Boundaries": "Boundaries"})
```

### Test

```python
clean_sf_df.head()
```

### Define - Problem 2

**2. Drop non-null values**

### Code

```python
clean_sf_df = clean_sf_df[clean_sf_df['Location End Date'].isnull()]
```

### Test

```python
clean_sf_df[clean_sf_df['Location End Date'].notnull()]
```

```python
clean_sf_df.info()
```

### Define - Problem 3

**3. Drop all duplicated rows.**

### Code

```python
clean_sf_df = clean_sf_df.drop_duplicates()
```

### Test

```python
# we will use the 'Fugazi Travel Agency, Inc.' example from earlier to check if it has been dropped
clean_sf_df[clean_sf_df['Name'] == 'Fugazi Travel Agency, Inc.']
```

```python
clean_sf_df[clean_sf_df.duplicated()]
```

### Define - Problem 4

**4. Remove additional digits in zip code if it exists**

### Code

```python
clean_sf_df['Zipcode'] = clean_sf_df['Zipcode'].astype(str)
for i in range(0, len(clean_sf_df)):
    clean_sf_df.iloc[i].Zipcode = clean_sf_df.iloc[i].Zipcode[:5]
```

### Test

```python
# we will use the tail end to check if it has been corrected
# we only need to check row 283367
clean_sf_df.tail()
```

### Define - Problem 5

**5. Convert datatypes to category.**

### Code

```python
# although this won't matter later, I just do it for practice
clean_sf_df['City'] = clean_sf_df['City'].astype('category')
clean_sf_df['Zipcode'] = clean_sf_df['Zipcode'].astype('category')
clean_sf_df['Industry'] = clean_sf_df['Industry'].astype('category')
clean_sf_df['Boundaries'] = clean_sf_df['Boundaries'].astype('category')
```

### Test

```python
clean_sf_df.info()
```

### Define - Problem 6

**6. Drop Location End Date.**

### Code

```python
clean_sf_df = clean_sf_df.drop(columns = ['Location End Date'])
```

### Test

```python
clean_sf_df.info()
```

### Define - Problem 7

**7. Drop all remaining null values.**

### Code

```python
clean_sf_df = clean_sf_df.dropna()
```

### Test

```python
clean_sf_df.info()
```

### Storing Data

```python
clean_sf_df.to_csv('bay_business_master.csv', index = False)
```

### Analysis, Insights & Visualization

```python
bay_master = pd.read_csv('bay_business_master.csv')
bay_master.head()
```

```python
bay_master.info()
```

#### Insight 1

```python
bay_master['Industry'].value_counts()
```

```python
from matplotlib.pyplot import figure

figure(figsize=(8, 6))
plt.rcParams.update({'font.size':12})
industry_count = bay_master['Industry'].value_counts().sort_values()

industry_count.plot(kind = 'barh')
plt.title("Industry Count In The Bay Area")
plt.ylabel('Industry Name')
plt.xlabel("Total Count");
```

**The top industy in SF/Bay Area is real estate, rental & leasing services then the professional, scientific and techinical services. Retail trade and food services rank 3rd and 4th respectively.**

#### Insight 2 - Top 25 Business Zip Code in Bay Area

```python
figure(figsize=(10, 8))
bay_master['Zipcode'] = bay_master['Zipcode'].astype('category')
zipcode_count = bay_master['Zipcode'].value_counts().sort_values(ascending = False).head(25)

zipcode_count.plot(kind = 'bar')
plt.title("Top 25 Business Zip Code in Bay Area")
plt.xticks(rotation = 80)
plt.xlabel('Zip Code')
plt.ylabel("Total Count");
```

**The most popular zip code is 94110, followed by 94103 then 94107.**

#### Insight 3 - Bar Chart Of Industry vs. Zip Code

```python
import seaborn as sb
figure(figsize=(25, 20))

sb.countplot(data = bay_master, x = 'Zipcode', hue = 'Industry',
            order = bay_master.Zipcode.value_counts().iloc[:10].index,
                   linewidth= 1,
                   edgecolor= 'black')
plt.legend(prop={'size': 12}, title = "Industries")
plt.title('Industry vs. Zip Code');
```

```python
figure(figsize=(25, 20))
plt.rcParams.update({'font.size':15})

sb.countplot(data = bay_master, x = 'Zipcode', hue = 'Industry',
            order = bay_master.Zipcode.value_counts().iloc[11:21].index,
                   linewidth= 1,
                   edgecolor= 'black')
plt.legend(bbox_to_anchor = (1, 1),prop={'size': 11}, title = "Industries")
plt.title('Industry vs. Zip Code');
```

**It seems like across all zip code the real estate, rental & leasing industry is the most popular industry.**

#### Insight 4 - Bar Chart Of Boundary Count

```python
bay_master['Boundaries'].value_counts().head(15)
```

```python
figure(figsize=(14, 8))
plt.rcParams.update({'font.size':10})
industry_count = bay_master['Boundaries'].value_counts().sort_values()

industry_count.plot(kind = 'barh')
plt.title("Boundary Count In SF")
plt.ylabel('Boundary Name')
plt.xlabel("Total Count");
```

**It should come as no surprise the top 3 boundaries with the most businesses are all located right next to one another where all the financial activities take place.**

#### Insight 5 - Boundaries vs. Industry - Top 20

**Zip code includes multiple boundaries, so we will analyze the boundaries to the industry to get a better idea what type of industries are packed in these smaller spaces.**

```python
figure(figsize=(25, 20))
plt.rcParams.update({'font.size':15})

sb.countplot(data = bay_master, x = 'Boundaries', hue = 'Industry',
            order = bay_master.Boundaries.value_counts().iloc[:10].index,
                   linewidth= 1,
                   edgecolor= 'black')
plt.xticks(rotation = 40)
plt.legend(bbox_to_anchor = (1, 1), prop={'size': 13}, title = "Boundaries")
plt.title('Boundaries vs. Industries: Top 10');
```

```python
figure(figsize=(25, 20))
plt.rcParams.update({'font.size':18})

sb.countplot(data = bay_master, x = 'Boundaries', hue = 'Industry',
            order = bay_master.Boundaries.value_counts().iloc[11:21].index,
                   linewidth= 1,
                   edgecolor= 'black')
plt.xticks(rotation = 40)
plt.legend(bbox_to_anchor = (1, 1),prop={'size': 13}, title = "Boundaries")
plt.title('Boundaries vs. Industries: 11-20');
```

**It seems the boundary with the most number of businesses is the Financial District/South Beach where professional, scientific and technical services are the most populat industry. This isn't necessarily a surprise, but it seems that real estate, rental & leasing services are the most popular in the other boundaries**

#### Insight 6 - Location Start Date Lineplot

```python
bay_master['Location Start Date'] = pd.to_datetime(bay_master['Location Start Date'], infer_datetime_format=True)
bay_master['year'] = pd.to_datetime(bay_master['Location Start Date']).dt.year
openyear_count = bay_master.groupby('year')['Location Start Date'].count()

figure(figsize=(25, 20))
plt.plot(openyear_count);
```

**Since the beginning, the number of business opened has experienced an exponential increase especially after the 1990s. You will notice a drastic drop in 2021. This is because there hasn't been enough data provided yet.**
